```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
bowlingdata = df[["Bowling_over","Bowling_runs","Bowling_wickets"]]
bowlingdata.fillna(0,inplace =True)
bowlingdata.head()
```

```python
def inertia_plotting(df):
    inertia = []
    K = range(1, 11)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(df)
        inertia.append(kmeans.inertia_)
    plt.figure(figsize=(10, 6))
    plt.plot(K, inertia, 'bo-')
    plt.xlabel('Number of Clusters K')
    plt.ylabel('Inertia')
    plt.title('Elbow Method to find Optimal K')
    plt.xticks(K)
    plt.grid()
    plt.show()
```

```python
batting_features = ['Batting_avg', 'Batting_strikerate', 'Batting_total_runs', 'Batting_boundary', 'Batting_dot']

# Select Bowling Features
bowling_features = ['Bowling_balls', 'Bowling_over', 'Bowling_runs', 'Bowling_economy', 
                    'Bowling_wickets', 'Bowling_avg', 'Bowling_strikerate']

# Separate batting and bowling data
batting_data = df[batting_features]
bowling_data = df[bowling_features]
```

```python
scaler = StandardScaler()
batting_data_scaled = scaler.fit_transform(batting_data)

# Standardize Bowling Data
bowling_data_scaled = scaler.fit_transform(bowling_data)
```

```python
inertia_plotting(batting_data_scaled)
```

```python
n_clusters = 3

# Batting Clustering
kmeans_batting = KMeans(n_clusters=n_clusters, random_state=42)
batting_clusters = kmeans_batting.fit_predict(batting_data_scaled)

# Bowling Clustering
kmeans_bowling = KMeans(n_clusters=n_clusters, random_state=42)
bowling_clusters = kmeans_bowling.fit_predict(bowling_data_scaled)
```
